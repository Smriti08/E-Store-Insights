```python
# top selling products in the superstore
product_group = df.groupby(['Sub-Category']).sum()['Sales']
product_group.head()
```

```python
top_selling_products = product_group.sort_values(ascending=False)
top_5_selling_products = pd.DataFrame(top_selling_products.head())
top_5_selling_products
```

```python
# Plotting Top 5 selling products in superstore
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
top_5_selling_products.plot(kind="bar")

# Add a title to the plot
plt.title("Top 5 Selling Products in Superstore")

# Add labels to the x and y axes
plt.xlabel("Sub-Category")
plt.ylabel("Total Profit")

# Show the plot
plt.show()
```

```python
# top-Discount products 

product_group = df.groupby(["Sub-Category"]).sum()["Discount"]

top_discount_products = product_group.sort_values(ascending=False)

top_5_discount_products =pd.DataFrame(top_discount_products[:5])
top_5_discount_products
```

```python
top_5_discount_products.plot(kind="bar")

plt.title("Top 5 Discount Products in Superstore")

plt.xlabel("Sub-Category")
plt.ylabel("Discount")

plt.show()
```

```python
# Are the top-selling products the most Discount?

top_5_discount_products.index == top_5_selling_products.index
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

# Plot the top 5 selling products in the first column
top_5_selling_products.plot(kind="bar", y="Sales", ax=ax1)

# Set the title for the first plot
ax1.set_title("Top 5 Selling Products")

# Plot the top 5 profit products in the second column
top_5_discount_products.plot(kind="bar", y="Discount", ax=ax2)

# Set the title for the second plot
ax2.set_title("Top 5 discount Products")

# Show the plot
plt.show()
```

```python
# Total profit and sales as per sub-category
df_cat.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.rcParams['figure.figsize']=[10,8]
plt.show()
```

```python
# sales and profit region wise
df_cat.groupby('Region')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Region')
plt.rcParams['figure.figsize']=[10,8]
plt.show()
```

```python
# heat map of correlation among the columns
fig,axes=plt.subplots(1,1,figsize=(9,6))
sns.heatmap(df_cat.corr(),annot=True)
plt.show()
```

```python
city_sales = df_cat.groupby('City', as_index=False).sum()

# Sort the data by Sales in descending order
city_sales.sort_values(by='Sales', ascending=False, inplace=True)

# Select the top 5 cities
top_5_cities_sales = city_sales.head()

plt.bar(top_5_cities_sales['City'], top_5_cities_sales['Sales'], align='center')
plt.xlabel("City")
plt.ylabel("Sales")
plt.title("Top 5 Cities by Sales")
plt.xticks(rotation=90)
plt.show()
top_5_cities_sales
```

```python
city_profit = df_cat.groupby('City', as_index=False).sum()

# Sort the data by Sales in descending order
city_profit.sort_values(by='Profit', ascending=False, inplace=True)

# Select the top 5 cities
top_5_cities_profit =city_profit.head()

plt.bar(top_5_cities_profit['City'], top_5_cities_profit['Profit'], align='center')
plt.xlabel("City")
plt.ylabel("Profit")
plt.title("Top 5 Cities by Profit")
plt.xticks(rotation=90)

plt.show()
top_5_cities_profit
```
